```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import os
import pymongo
```

```python
url1 = 'https://mars.nasa.gov/news/8915/nasas-mars-helicopter-to-make-first-flight-attempt-sunday/'
```

```python
url1
```

```python
response = requests.get(url1)
mars = BeautifulSoup(response.text, 'html.parser')
```

```python
print(mars.prettify())
```

```python
results = mars.find_all('section', class_="content_page module")
results
```

```python
for result in results:
    # Error handling
    try:
        # Identify and return title of listing
        news_title = result.find('h1', class_='article_title').text
        news_p = result.find_all("p")[2].text
        news_date = result.find('span', class_='release_date').text.split('|')[1]
        print('\n--------\n')
        print("Title:", news_title)
        print("News:", news_p, "\n")
        print("Release Date:", news_date)
        print('\n--------\n')
    except AttributeError as e:
        print(e)
```

```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url12 = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
```

```python
response12 = requests.get(url12)
mars12 = BeautifulSoup(response12.text, 'html.parser')
```

```python
import re
results12 = mars12.find_all('a', {'href':re.compile('/search/map/Mars')})
img_u=[]
for result12 in results12:
    try:
        img_u.append('https://astrogeology.usgs.gov' + result12['href'])
        #print('--------')
        #print("Image_u:", img_u)
    except AttributeError as e:
        print(e)
```

```python
img_u
```

```python
hemisphere_image_urls=[]

for i in range(len(img_u)):
    url_i = img_u[i]
    browser.visit(url_i)
    xpath_i = './/ul//a[@href]'
    results_i = browser.find_by_xpath(xpath_i)
    img = results_i[0]
    img.click()    
    
    response_i = requests.get(url_i)
    mars_i = BeautifulSoup(response_i.text, 'html.parser')
    results_ii=mars_i.find_all('a', {'href':re.compile('full.jpg')})
    #results_ti=mars_i.find_all('h2', {'class':re.compile('title')})

    for result_ii in results_ii:
        try:
            title=mars_i.find('h2', class_='title').text.split(' Enhanced')[0]
            img_url=result_ii['href']
            hemisphere_image_urls.append({"title": title, "img_url": img_url})
            print('--------')
            print("title:", title)
            print("img_url:", img_url)
        except AttributeError as e:
            print(e)
```

```python
hemisphere_image_urls 
```

```python
browser.quit()
```
